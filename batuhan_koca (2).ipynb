## filtering flights which flies more than 80 minutes and delayed 

```r
flights [ air_time > 80 & arr_delay > 0 ]
```

## J operation. Showing only arrive delays and the speeds of planes type of miles per hour.

```r
flights [, .(arr_delay, speedmilehour = distance / ((air_time %/% 100) + ((air_time %% 100)/60)))]
```

## J operation. Adding speeds, kilometers per hour.

```r
flights [, speedkmh:= (distance * 1.609344) / ((air_time %/% 100) + ((air_time %% 100)/60))]
flights [, .(speedkmh)]
flights
```

## By operation. Adding minimum and maximum department time of every day in the year. 

```r
flights [, .(min_dep_time = min(dep_time, na.rm=T), max_dep_time = max(dep_time, na.rm=T)), by = c("month","day") ]
```
